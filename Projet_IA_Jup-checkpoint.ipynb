# Régression linéaire simple (sans sklearn)

Importation et visualisation des données.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# importation des données
data=pd.read_csv('./Data_Reg/reg_simple.csv')
print(data)
```

Initialisation des graphes et valeurs

```python
x0=(data.note)
x1=np.ones(len(data))
x=np.column_stack((x0,x1))
y=np.array(data.heure_rev)
y=y.reshape((len(y),1))

# Génération aléatoire de a et b
theta = np.array(np.random.rand(x.shape[1], 1))
print('Theta original :',theta)
y.shape
print(len(y))
```

```python
plt.figure()
plt.scatter(data.note,data.heure_rev)
plt.xlabel('Heure_rev')
plt.ylabel('Note')


#Définition du modèle
def model(x,theta):
    f=np.dot(x,theta)
    return f
   
#Fonction coût
def fonction_cout(x,y,theta):
    m = len(y)
    return (1/(2*m)) * np.sum((model(x, theta) - y)**2)

#Calcul du gradient
def grad(x,y,theta):
    dJ=(1/len(y))*(x.T.dot(model(x,theta)-y))
     
    return(dJ)

#Descente de gradient
def descente_gradient(x,y,theta,alpha,n_iterations):
    cout = np.zeros(n_iterations)
    plt.figure()
    for i in range (0,n_iterations):
        # f=model(x,theta);
        # if i%10==0:
        #     plt.plot(x, model(x,theta), lw=0.5)
        cout[i] = fonction_cout(x, y, theta)
        theta = theta - alpha * grad(x,y,theta)
    return(theta, cout)

n_iterations = 40
learning_rate = 0.0001
 
theta_final, cost_history = descente_gradient(x, y, theta, learning_rate, n_iterations)

# Voici les parametres du modele une fois que la machine a été entrainée
print(theta_final) 
 
# Création d'un vecteur prédictions qui contient les prédictions de notre modele final
predictions = model(x, theta_final)
 
# Affiche les résultats de prédictions (en rouge) par rapport a notre Dataset (en bleu)
plt.scatter(x0, y)
plt.plot(x0, predictions, c='r')
plt.show()
    
# Affichage de la fonction prédite
print("La fonction finale est : y=",theta_final[0].round(2),"x+",theta_final[1].round(2))

#Affichage de la fonction coût en fonction du nombtre d'itérations
plt.plot(range(n_iterations), cost_history)
plt.show()
```

```python

#Evaluer votre modèle en utilisant le coefficient de détermination

R_2=1-(sum(y-predictions)**2)/(sum((y-np.mean(y))**2))
print(R_2)
```

```python
#Visualising the training set result
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model_linreg.predict(X_train), color = 'blue')
plt.show
```

```python
from sklearn.metrics import r2_score
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, model_linreg.predict(X_test), color = 'blue')
plt.show
sc=r2_score(y_test, model_linreg.predict(X_test))
print("Score= ",sc)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
sc=r2_score(y_test,y_pred)
print("Score= ",sc)
print('mean_squared_error =',mse,'\nmean_absolute_error =', mae)
```
